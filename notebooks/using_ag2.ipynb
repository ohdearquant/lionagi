```python
# load environment variables
from dotenv import load_dotenv
load_dotenv()
```

```python
# disable logging
import logging
logging.getLogger("httpx").setLevel(logging.WARNING)
logging.getLogger("autogen.oai.client").setLevel(logging.ERROR)
```

```python
llm_config = {
    "api_type": "openai",
    "model": "gpt-5.4-mini",
}
    
agent_configs = [
    {
        "name": "Researcher",
        "role": "data gatherer",
        "system_message": (
            "You research the topic and provide 2-3 key facts. "
            "Be concise — max 3 sentences per turn."
        ),
        "tools": [],
        "handoffs": [
            {
                "target": "Analyst",
                "condition": "When you have gathered initial facts",
            }
        ],
    },
    {
        "name": "Analyst",
        "role": "synthesizer",
        "system_message": (
            "You synthesize findings into a brief insight. "
            "Max 3 sentences. When done, say TERMINATE."
        ),
        "tools": [],
        "handoffs": [],
    },
]
```

```python
import lionagi as li

model = li.iModel(
    provider="ag2",
    agent_configs=agent_configs,
    llm_config=llm_config,
)

class Trend(li.BaseModel):
    name: str
    description: str
    reason: str

async with li.Branch(chat_model=model) as b:
    result = await b.operate(
        instruction="What were the key trends in AI hiring in 2024?",
        field_models=[li.Spec(list[Trend], name="trends")],
    )
    for i, trend in enumerate(result.trends):
        print(f"  {i + 1}. {trend.name}: {trend.description} (because {trend.reason})\n")
```
